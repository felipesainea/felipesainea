```python
import pandas as pd


# Descargar el conjunto de datos
#path = 'h134fpGtR6M_yvc-2I_et9uCRs-ix551GcDDHNh0tI4tw'
path = "https://docs.google.com/spreadsheets/d/134fpGtR6M_yvc-2I_et9uCRs-ix551GcDDHNh0tI4tw/export?format=xlsx"
df = pd.read_excel(path)

# Obtener información sobre la cantidad de variables y registros
num_variables = len(df.columns)
num_registros = len(df)
print(f'\nCantidad de variables: {num_variables}')
print(f'Cantidad de registros: {num_registros}')
```

```python
# Identificar variables numéricas y cualitativas
variables_numericas = df.select_dtypes(include=['float64', 'int64']).columns
variables_cualitativas = df.select_dtypes(include=['object']).columns

print('\nVariables Numéricas:')
print(variables_numericas)
print('\nVariables Cualitativas:')
print(variables_cualitativas)

# Resumen de las variables numéricas
print('\nResumen de Variables Numéricas:')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier


# Definir variables predictoras (X) y variable objetivo (y)
X = df.drop('NObeyesdad', axis=1)  # Variables predictoras
y = df['NObeyesdad']  # Variable objetivo

# Codificación one-hot de las variables categóricas en el conjunto de datos
X_encoded = pd.get_dummies(X)

# División de los datos en entrenamiento y testeo (70% para entrenamiento, 30% para testeo)
X_train_encoded, X_test_encoded, y_train, y_test = train_test_split(X_encoded, y, test_size=0.3, random_state=42)

# Crear modelos
log_reg_model = LogisticRegression(max_iter=50000)
rfc_model = RandomForestClassifier()
decision_tree_model = DecisionTreeClassifier(max_depth=5, min_samples_split=2, min_samples_leaf=1)


# Entrenar modelos
log_reg_model.fit(X_train_encoded, y_train)
rfc_model.fit(X_train_encoded, y_train)
decision_tree_model.fit(X_train_encoded, y_train)

# Realizar predicciones
y_pred_log_reg = log_reg_model.predict(X_test_encoded)
y_pred_rfc = rfc_model.predict(X_test_encoded)
y_pred_decision_tree = decision_tree_model.predict(X_test_encoded)


# Reportes de clasificación
print("Reporte de Clasificación para Regresión Logística:")
print(classification_report(y_test, y_pred_log_reg))

print("\nReporte de Clasificación para Random Forest Classifier:")
print(classification_report(y_test, y_pred_rfc))

print("\nReporte de Clasificación para Árbol de Decisión:")
print(classification_report(y_test, y_pred_decision_tree))
```

#Recomendaciones para mejorar:

Reentrenamiento y Ajuste de Parámetros: Para los modelos con menor rendimiento como el Árbol de Decisión, considerar el ajuste de parámetros como la profundidad máxima y el número mínimo de muestras por hoja podría mejorar la precisión y la generalización.

# Reentrenamiento

```python
from sklearn.model_selection import GridSearchCV

# Modelo
decision_tree_model = DecisionTreeClassifier()

# Definir la grilla de parámetros a buscar
param_grid = {
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 10, 20],
    'min_samples_leaf': [1, 5, 10]
}

# Configurar GridSearchCV
grid_search = GridSearchCV(decision_tree_model, param_grid, cv=5, scoring='accuracy')

# Entrenamiento utilizando GridSearchCV
grid_search.fit(X_train_encoded, y_train)

# Mejores parámetros encontrados
print("Mejores parámetros:", grid_search.best_params_)

# Mejor modelo encontrado por GridSearchCV
best_model = grid_search.best_estimator_

# Evaluación
y_pred_decision_tree = best_model.predict(X_test_encoded)
print("Reporte de Clasificación para Árbol de Decisión Mejorado:")
print(classification_report(y_test, y_pred_decision_tree))
```

##Explicación de la Validación

>GridSearchCV: Utiliza validación cruzada, dividiendo el conjunto de datos de entrenamiento en k subconjuntos más pequeños (en este caso, cv=5). Entrena el modelo en k-1 de estos subconjuntos y valida el modelo en el subconjunto restante, repitiendo el proceso 'k' veces con un subconjunto diferente cada vez. Esto ayuda a asegurar que el modelo sea robusto y generalice bien a nuevos datos.

# Conclusiones sobre el Reentrenamiento

Tras aplicar diversos modelos de clasificación al conjunto de datos y realizar ajustes mediante el reentrenamiento, los resultados han demostrado que el Árbol de Decisión mejorado, con los parámetros ajustados (max_depth: 10, min_samples_leaf: 1, min_samples_split: 2), ha proporcionado un rendimiento sobresaliente. Este modelo alcanzó una precisión general del 92%, con notables mejoras en todas las categorías comparado con los resultados iniciales y otros modelos como la Regresión Logística y el Random Forest. El ajuste de parámetros permitió al Árbol de Decisión manejar mejor la complejidad y las variaciones del conjunto de datos, equilibrando adecuadamente entre el sobreajuste y el subajuste, como se refleja en las altas puntuaciones de precisión y recall en todas las categorías, especialmente en tipos de obesidad donde alcanzó casi la perfección. Estos resultados resaltan la efectividad del Árbol de Decisión mejorado en el manejo de datos complejos y su capacidad para generalizar bien sobre datos nuevos, haciéndolo el modelo más adecuado entre los evaluados para esta tarea específica.
